```python
import pandas as pd
import numpy as np
```

```python
#mlb_pitches_2015_to_2018 = pd.read_csv("mlb_pitches_2015_to_2018.csv")
#games = pd.read_csv("games.csv")
pitches = pd.read_csv("pitches.csv")
```

```python
player_names = pd.read_csv("player_names.csv")
```

```python
atbats = pd.read_csv("atbats.csv")
```

```python
atbats = atbats.rename(columns={'stand': 'batter_side'})
```

```python
pitcher_throwing_hand = atbats[["p_throws", "pitcher_id", "ab_id", "batter_side"]]
pitcher_throwing_hand = pitcher_throwing_hand.drop_duplicates()
```

```python
pitches
```

```python
player_names
```

```python
pitcher_name_and_id = pitcher_throwing_hand.merge(player_names, left_on="pitcher_id", right_on="id", how="inner")
pitcher_name_and_id["name"] = pitcher_name_and_id["first_name"] + pitcher_name_and_id["last_name"]
pitcher_name_and_id = pitcher_name_and_id.drop(columns = ["first_name", "last_name", "id"])
pitcher_name_and_id["name"] = pitcher_name_and_id["name"].str.lower()
pitcher_name_and_id
```

```python
pitches = pitches.dropna(how='any',axis=0) 
```

```python
df_final = pitches.merge(pitcher_name_and_id, on="ab_id")
df_final
```

```python
df_final[df_final["ab_id"] == 2015000001]
```

```python
df_final.isnull().sum(axis = 0)
```

```python
df_final = df_final[["px", "pz", "start_speed", "end_speed", 
         "spin_rate", "spin_dir", "break_angle", 
        "break_length", "break_y", "ax", "ay", "az", 'p_throws',
        "sz_bot", "sz_top", 'pitcher_id', "batter_side",
        "vx0", "vy0", "vz0", "x", "x0", "y", "y0",
        "z0", "pfx_x", "pfx_z", "nasty", "zone", 
        "code", "type", "pitch_type", "b_count",
        "s_count", "pitch_num"]]
```

```python
df_final = df_final.astype({'p_throws': 'string', 'code' : "string", "type" : "string" , "pitch_type" : "string"})
```

```python
df_final = df_final[df_final["pitch_type"] != "PO"]
df_final = df_final[df_final["pitch_type"] != "FA"]
df_final = df_final[df_final["pitch_type"] != "SC"]
df_final = df_final[df_final["pitch_type"] != "FO"]
df_final = df_final[df_final["pitch_type"] != "UN"]
df_final = df_final[df_final["pitch_type"] != "AB"]
df_final = df_final[df_final["pitch_type"] != "IN"]
df_final = df_final[df_final["pitch_type"] != "EP"]
```
